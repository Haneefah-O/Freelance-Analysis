```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
max_weather = pd.read_csv("Weather Data/maxtemp_annual_timeseries.csv")
max_weather.info()
```

```python
max_weather.head()
```

```python
min_weather = pd.read_csv("Weather Data/Mintemp_annual_timeseries_EDIT.csv")
min_weather.info()
```

```python
plt.plot(max_weather.Year, max_weather.Nigeria)
```

```python
min_weather.plot(x='Year', y='Nigeria', kind='line') 
```

```python
plt.plot(max_weather.Year, max_weather.Nigeria, min_weather.Nigeria)
```

```python
max_min_weather = max_weather.merge(min_weather, on='Year', suffixes=('_max', '_min'))

max_min_weather.head()
```

```python
max_min_weather.plot(x='Year', y=['Nigeria_max', 'Nigeria_min'])
plt.ylabel('Temperature')
```

```python
mean_temp = pd.read_csv("Weather Data/Mean_Temperature_Timeseries.csv")
mean_temp.head()
```

```python
mean_temp.drop('Nigeria', axis=1 )
```

```python
weather_temp_stat = max_min_weather.merge(mean_temp, on='Year')
weather_temp_stat.plot(x='Year', y=['Nigeria_max', 'Nigeria_min', 'Nigeria_mean'], label=['Min', 'Max', 'Average'])
plt.ylabel('Temparature')
plt.title('Trend in Weather in Nigeria')
```

```python
temp_50 = weather_temp_stat.iloc[70:121, :]
temp_50.head()
```

```python
temp_50.plot(x='Year', y=['Nigeria_max', 'Nigeria_min', 'Nigeria_mean'], label=['Min', 'Max', 'Average'])
plt.ylabel('Temparature')
plt.title('Weather Trends in Nigeria')
```

```python
max_min_weather.plot(x='Year', y=['Lagos_max', 'Akwa Ibom_max', 'Benue_max'], label=['Lagos', 'Akwa Ibom', 'Benue'])
plt.ylabel('Temparature')
plt.title('High Temperatues in Southern Region')
```

```python
max_min_weather.plot(x='Year', y=['Lagos_min', 'Akwa Ibom_min', 'Benue_min'], label=['Lagos', 'Akwa Ibom', 'Benue'])
plt.ylabel('Temparature')
plt.title('Low Temperatues in Southern Region')
```

```python
max_min_weather.plot(x='Year', y=['Sokoto_max', 'Borno_max', 'Kaduna_max'], label=['Sokoto', 'Borno', 'Kaduna'])
plt.ylabel('Temparature')
plt.title('High Temperatues in Northern Region')
```

```python
max_min_weather.plot(x='Year', y=['Sokoto_min', 'Borno_min', 'Kaduna_min'], label=['Sokoto', 'Borno', 'Kaduna'])
plt.ylabel('Temparature')
plt.title('Low Temperatues in Northern Region')
```
